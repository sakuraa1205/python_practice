# 3. 판다스 데이터프레임 조회

1) 데이터 읽어오기

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/Attrition_simple2.CSV'
data = pd.read_csv(path)  

# 상위 5개 확인
data.head()
```

2) 데이터 열 조회

```python
# 데이터 프레임 : 2차원 구조
# 시리즈 : 1차원 구조
# 시리즈로 조회
a = (data['Age'])
print(a)
print(type(a))
```

```python
# 데이터 프레임으로 조회
# 대괄호 안에 리스트!
b = data[['Age', 'Gender']]
print(b)
print(type(b))
```

3) 조건으로 조회

```python
# 데이터 확인
data.head()
```

```python
# 조건문의 결과는 True | False
data['Age'] > 30
```

```python
# 값으로 받기
# loc 쓰는게 권장된다!!
data.loc[data['Age'] > 30]
```

```python
# 여러 조건 조회
data.loc[(data['Age'] > 30) & (data['Age'] < 40)]
```

```python
# or 조건
data.loc[(data['Age'] > 30) | (data['Age'] < 40)]
```

```python
# isin(), between() 메소드
# 편리함
# isin() = or = | 조건이랑 같음
# between() = and = & 조건이랑 같음 (사이값)
display(data.loc[data['JobSatisfaction'].isin([1, 4])])
print('*' * 135)
display(data.loc[data['JobSatisfaction'].between(1, 4)])
```

```python
# 조건에 맞는 하나의 열 조회
data.loc[data['JobSatisfaction'] == 1][['Age']]
```

```python
# 조건에 맞는 여러 열 조회
data.loc[data['JobSatisfaction'] == 1][['DistanceFromHome', 'Age', 'MonthlyIncome']]
```

# 4. 판다스 데이터 프레임 집계

1) 데이터 프레임 집계

```python
# 열 하나 조건
data['Age'].sum()
```

```python
# 여러 열 조건
data[['JobSatisfaction', 'Age']].mean()
```

```python
# 하나의 열 집계 : ~ 별로 조건을 구함
# data.groupby('~', as_index = True or False)
print(data.groupby('JobSatisfaction', as_index = True)['Age'].mean())
print(data.groupby('JobSatisfaction', as_index = False)['Age'].mean())
```

```python
# 여러개의 열 집계 : ~ 별로 ~, ~ 조건을 구함
print(data.groupby('JobSatisfaction', as_index = True)[['Age', 'MonthlyIncome']].mean())
print(data.groupby('JobSatisfaction', as_index = False)[['Age', 'MonthlyIncome']].mean())
```

```python
# 여러 함수로 한꺼번에 집계
data.groupby('JobSatisfaction', as_index = True)[['Age', 'MonthlyIncome']].agg(['min', 'max', 'std', 'mean'])
```

# 5. 판다스 데이터 프레임 변경

1) 데이터 프레임 변경

```python
# 데이터 프레임 확인
data.head()
# 선택한 열 이름 변경
data.rename(columns = {'DistanceFromHorm' : 'Distance',
                      'MonthlyIncome' : 'M_Income', 
                      'PercentSalaryHike' : 'P.S.H'}, inplace = True)
```

```python
# 변경 된 데이터 프레임 확인
data.head()
```

```python
# 모든 열 이름 변경
data.columns = ['Attr','Age','Dist','EmpNo','Gen','JobSat','Marital','M_Income', 'OT', 'PctSalHike', 'TotWY']

data.head()
```

2) 데이터 프레임 열 추가

```python
data['Income_LY'] = data['M_Income'] / (1 + data['PctSalHike'] / 100)
data.head()
```

3) 데이터 프레임 열 삭제

```python
# 삭제는 항상 조심해야 함!
# axis = 0 행, axis = 1 열
# np 에서는 axis = 0은 행 방향, axis = 1은 열 방향
# 한 개의 열 삭제
data.drop('Income_LY', axis = 1, inplace = True)
data.head()
```

```python
# 여러개의 열 삭제
data.drop(['TotWY', 'OT'], axis = 1, inplace = True)
data.head()
```

4) 유익한 기능

```python
# map method!
data['Gen'] = data['Gen'].map({'Male' : 1, 'Female' : 0})
data.head()
```

```python
# cut 함수
# 범위를 3등분해서 등급 구하기
pd.cut(data['M_Income'], 3)
```

```python
# 범위 3등분해서 등급 매기기
data['Income_Grade'] = pd.cut(data['M_Income'], 3, labels = ['a', 'b', 'c'])
data.head()
```

```python
# 수입 등급 별 수입의 최소값 최대값 표준편차를 구하라.
data.groupby('Income_Grade', as_index = True)['M_Income'].agg(['min', 'max', 'std'])
```
