# 결합하기(merge)

```python
# 데이터 불러오기
sales = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/sales.csv")
products = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/products.csv")
```

```python
# 데이터 확인
display(sales.head())
display(products.head())
```

```python
# 데이터 결합하기
display(pd.merge(sales, products, how = 'left').head())
display(pd.merge(sales, products, how = 'right').head())
display(pd.merge(sales, products, how = 'outer').head())
```

# Merge 실습해보기

```python
data = pd.merge(sales, products)
data.head()
```

```python
# Category 별 총 매출액
data.groupby('Category', as_index = False)[['Amt']].sum()
```

```python
# Category 별 총 판매량
data.groupby('Category', as_index = False)[['Qty']].sum()
```

```python
# 전체에서 가장 매출이 높은 상품 10개 조회
df1 = data.groupby('ProductName', as_index = False)[['Amt']].sum()
df1.sort_values(by = 'Amt', ascending = False).head(10)
```

# 결합하기(concat)

```python
# 위아래로 붙이기
pd.concat([sales, products])
```

```python
# 옆으로 붙이기
pd.concat([sales, products], axis = 1)
```

# Rolling & Shift

```python
# 데이터 불러오기
stock = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/SK.csv') 
stock.head()
```

```python
# 전날 종가, 전전날 종가
stock['a day ago'] = stock['Close'].shift()
stock['two days ago'] = stock['Close'].shift(2)
stock.head()
```

```python
# 전전날 대비 전날 종가 차이
stock['Price Change'] = stock['a day ago'] - stock['two days ago']
stock.head()
```

```python
# 시초가, 당일 최대가, 당일 최저가 등 Column 제거
stock.drop(['Open', 'High', 'Low', 'AdjClose'], axis = 1)
```

```python
# 3일 최대 주가
stock['3Day_Max_1'] = stock['Close'].rolling(3).max()
stock.head()
```

```python
# 기준일 포함 x, 3일 최대 주가
stock['3Day_Max_2'] = stock['Close'].rolling(3).max().shift()
stock.head()
```

```python
# 기준일 포함 x, 7일 평균 거래량
stock['7Day_Mean_Vol'] = stock['Volume'].rolling(7).mean().shift()
stock.head(10)
```

```python
# 기준일 포함 x, 7일 평균 거래량, 1일만 있어도 표현 가능하게
stock['7Day_Mean_Vol'] = stock['Volume'].rolling(7, min_periods = 1).mean().shift()
stock.head(10)
```
